```python
import pandas as pd
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df=pd.read_csv("fremont-bridge.csv", index_col= 'Date')
df.head()
```

```python
#"Total" sütunu ekleyerek bu veri kümesini oluşturalım
```

```python
df.columns=["West", "East"]
df['Total']=df["West"]+df[ "East"]
```

```python
df.head()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df=df.fillna(df.mean())
#boş verileri ortalama ile doldurduk
```

```python
df.isnull().sum()
```

```python
figure=px.line(df,y="Total",title="Hourly Bicycle count")

figure.show()
```
